#### 1.1.) Обоснование выбора распределения

Пропорция техники с возможностью подключения к "Умному дому" и без такой возможности составляет 1:3, общая "часть" равна $1+3=4$.

Вероятность выбора экземпляра с "Умным домом" $P_{уд}$ равна $1/4=0.25$.
Вероятность выбора экземпляра без "Умного дома" $(1 - P_{уд})$ равна $3/4=0.75$.

У нас есть 7 независимых испытаний (выбор 7 экземпляров техники для розыгрыша), и каждое из них имеет только два исхода (с возможностью подключения к системе "Умный дом" и без нее). Распределение случайной велчины, принимающей только два значения, соответствует распределению Бернулли. В нашем случае, происходит повторение нескольких независимых испытаний Бернулли, то есть происходит процесс Бернулли. А значит, результаты процесса Бернулли будут следовать **биномиальному распределению**. (Биномиальным называют распределение, при котором возможны только два исхода (успех или неудача, выигрыш или проигрыш) и вероятность успеха и неудачи одинакова для всех испытаний. Однако исходы не обязательно должны быть равновероятными, и каждое испытание не зависит от других.)

Нам нужно найти вероятность того, что среди 7 выбранных экземпляров техники будет не более двух с возможностью подключения к "Умному дому". Это означает, что среди семи экземпляров, с возможностью подключения к "Умному дому", может быть 0 или 1, или 2.

Обозначим искомую вероятность как:

$P_{уд}(X \leq 2)$

Она будет равнятся сумме вероятностей, интересующих нас исходов:

$P_{уд}(X \leq 2) = P_{уд}(X = 0) + P_{уд}(X = 1) + P_{уд}(X = 2)$

Для того, чтобы оценить вероятность, что среди $n$ испытаний будет $k$ успехов, используют следующую формулу:

$P(X = k) = \begin{pmatrix} n \\ k \end{pmatrix} p^k (1-p)^{n-k}$, где

$n = 7$ - общее количество испытаний,

$p = P_{уд} = 0.25$ - вероятность успеха в каждом испытании,

$k$ - количество успехов, в нашем случае поочередно будет принимать значения 0,1,2,

$\begin{pmatrix} n \\ k \end{pmatrix}$ - называют биномиальным коэффициентом, и он вычисляется следующим образом:

$\frac{n!}{k! (n-k)!}$

#### 1.2.) Расчет вероятности

Расчитаем вероятность, что среди 7 экземпляров будет 0 с возможностью подключения к "Умному дому":

$P_{уд}(X = 0) = \begin{pmatrix} 7 \\ 0 \end{pmatrix} 0.25^0 (1-0.25)^{7-0} = \frac{7!}{0! (7-0)!} * 1 * 0.75^7 = \frac{5040}{1 * 5040} * 1 * 0.13348388671875 = 0.13348388671875$

Расчитаем вероятность, что среди 7 экземпляров будет 1 с возможностью подключения к "Умному дому":

$P_{уд}(X = 1) = \begin{pmatrix} 7 \\ 1 \end{pmatrix} 0.25^1 (1-0.25)^{7-1} = \frac{7!}{1! (7-1)!} * 0.25 * 0.75^6 = \frac{5040}{1 * 720} * 0.25 * 0.177978515625 = 0.31146240234375$

Расчитаем вероятность, что среди 7 экземпляров будет 2 с возможностью подключения к "Умному дому":

$P_{уд}(X = 2) = \begin{pmatrix} 7 \\ 2 \end{pmatrix} 0.25^2 (1-0.25)^{7-2} = \frac{7!}{2! (7-2)!} * 0.0625 * 0.75^5 = \frac{5040}{2 * 120} * 0.0625 * 0.2373046875 = 0.31146240234375$

Расчитаем вероятность, что среди 7 экземляров будет не более двух с возможностью подключения к "Умному дому":

$P_{уд}(X \leq 2) = P_{уд}(X = 0) + P_{уд}(X = 1) + P_{уд}(X = 2) = 0.13348388671875 + 0.31146240234375 + 0.31146240234375 = 0.75640869140625$

Округлим искомую вероятность до 3-х знаков после запятой $P_{уд}(X \leq 2) = 0.756$

#### 1.3.) Верный итоговый ответ

Ответ: $P_{уд}(X \leq 2) = 0.756$

#### 2.1.) Нахождение пропущенной вероятности

Сумма вероятностей всех возможных значений случайной величины всегда равна единице. Следовательно:

$P(X = 0) = 1 - P(X = 1) - P(X = 2) - P(X = 3) - P(X = 4) - P(X = 5) = 1 - 0.3 - 0.3 - 0.2 - 0.1 - 0.1 = 0$


$P(X = 0) = 0$

$$
\begin{array}{|c|c|c|c|c|c|c|}
\hline X \text { (число публикаций) } & 0 & 1 & 2 & 3 & 4 & 5 \\
\hline \mathrm{P} \text { (вероятность) } & 0 & 0,3 & 0,3 & 0,2 & 0,1 & 0,1 \\
\hline
\end{array}
$$

#### 2.2.) Вычисление математического ожидания

Математическое ожидание случайной величины — это взвешенное среднее значений этой величины с учётом вероятностей. Оно вычисляется по следующей формуле:

$E(X) = \mu_X = \sum x_i \cdot P(X = x_i)$, где

$i$ - индекс переменной,

$x_i$ - значение случайной величины с индексом $i$,

$P(X = x_i)$ = вероятность того, что случайная величина принимает значение $x_i$.

$E(X) = x_0 * P(X = 0) + x_1 * P(X = 1) + x_2 * P(X = 2) + x_3 * P(X = 3) + x_4 * P(X = 4) + x_5 * P(X = 5) = 0*0 + 1*0.3 + 2*0.3 + 3*0.2 + 4*0.1 + 5*0.1 = 0 + 0.3 + 0.6 + 0.6 + 0.4 + 0.5 = 2.4$

$E(X) = 2.4$ 

#### 2.3.) Вычисление дисперсии

Дисперсия — это мера, которая показывает, насколько широко или узко разбросаны данные вокруг их среднего значения (математического ожидания). Она вычисляется по следующей формуле:

$V(\mathrm{X})=D(\mathrm{X})=\sigma_{X}^{2}=E\left(\mathrm{X}^{2}\right)-(E(\mathrm{X}))^{2}$,

Вычислим дисперсию для нашего случая:

$V(\mathrm{X})=D(\mathrm{X})=\sigma_{X}^{2} = E(X^2) - (E(X))^2 = ({x_0}^2 * P(X = 0) + {x_1}^2 * P(X = 1) + {x_2}^2 * P(X = 2) + {x_3}^2 * P(X = 3) + {x_4}^2 * P(X = 4) + {x_5}^2 * P(X = 5)) - (2.4)^2 = (0*0 + 1*0.3 + 4*0.3 + 9*0.2 + 16*0.1 + 25*0.1) - 5.76 = 0 + 0.3 + 1.2 + 1.8 + 1.6 + 2.5 - 5.76 = 7.4 - 5.76 = 1.64$

Вычислим стандартное отклонение: 

$\sigma_{X} = \sqrt{V(\mathrm{X})} = \sqrt{1.64} = 1.281$

$V(\mathrm{X}) = 1.64$

$\sigma_{X} = 1.281$

#### 2.4.) Решение с использованием Python



```python
import pandas as pd
import numpy as np

# Закон распределния вероятностей количества публикаций
data = {
    'X(число публикаций)': [0, 1, 2, 3, 4, 5],
    'P(вероятность)': [np.nan, 0.3, 0.3, 0.2, 0.1, 0.1] 
}
df = pd.DataFrame(data)
print("Исходный DataFrame:")
print(df)

# Вычисление пропущенной вероятности
sum_of_known = df['P(вероятность)'].sum()
missing_prob = 1.0 - sum_of_known
df['P(вероятность)'] = df['P(вероятность)'].fillna(missing_prob)

print("\nDataFrame после вычисления:")
print(df)
print(f"\nПропущенная вероятность равна: {missing_prob}")

# Вычисление математического ожидания
E_x = (df['X(число публикаций)'] * df['P(вероятность)']).sum()
print(f"\nМатематическое ожидание равно: {E_x}")

# Вычисляем дисперсию
df['X(число публикаций)^2'] = df['X(число публикаций)'] ** 2
E_X2 = (df['X(число публикаций)^2'] * df['P(вероятность)']).sum()

V_x = E_X2 - (E_x ** 2)

print(f"\nДисперсия равна: {round(V_x, 2)}")
print(f"\nСтандартное отклонение равно: {round((round(V_x, 2)**0.5), 3)}")

```

#### 3.1.) Корректное решение первой подзадачи 
Вычислим вереятность того, что за месяц сломается ровно 1 ноутбук.
В нашем случае есть событие (поломка ноутбука), дана вероятность этого события за определенный период времени (месяц), что соответсвует распределению Пуассона. (Распределение Пуассона — это дискретное распределение вероятностей числа событий, происходящих в данный период времени, с учётом среднего количества раз, когда событие происходит за этот период времени.)

В распределении Пуассона значение случайной величины может быть любым неотрицательным числом.  Случайная величина будет обладать следующими характеристиками:

$EX  = \lambda$ - математическое ожидание

$\sigma_X = \sqrt{\lambda}$ стандартное отклонение,

где  — $\lambda$ ожидаемое число событий за период времени.

Чтобы рассчитать вероятность того, что за период времени произойдёт  $k$ событий, можно пользоваться следующей формулой:

$P(X = k) = \frac{\lambda^k e^{- \lambda}}{k!}$, где

$k = 1$ - количество сломанных ноутбуков,

$\lambda = n * p = 10000*0.0002 = 2$ - ожидаемое количество поломок ноутбуков за месяц, где

$n = 10000$ - количество ноутбуков,

$p = 0.0002$ - вероятность поломки одного ноутбука

$P(X = 1) = \frac{2^1 * 2.71828^{- 2}}{1!} = \frac{2 * 0.13533546530402737}{1} = \frac{0.27067093060805475}{1} = 0.27067093060805475$

$P(X = 1) = 0.27067093060805475$

Также Вычислим вереятность того, что за месяц сломается ровно 1 ноутбук с помощью Python.



```python
import scipy
import seaborn as sns
from numpy import random

# ожидаемое количество поломок ноутбуков за месяц
lmd = 10000*0.0002

print(f"\nВероятность поломки одного ноутбука равна: {scipy.stats.distributions.poisson.pmf(1, lmd)}")

ax = sns.displot(random.poisson(lam=lmd, size=10000), kde=False)
ax.set(xlabel='Значение случайной величины', ylabel='Частота')
```

#### 3.2.) Корректное решение второй подзадачи пункта а
Вычислим вероятность того, что за месяц сломается более 5 ноутбуков. 
Общее количество ноутбуков 10 000. Следовательно нам необходимо найти вероятность того, что сломается от 6 до 10 000 ноутбуков $P(5 > X \leq 10000)$. Это значит, что необходимо найти сумму вероятностей поломки каждого количества ноутбуков из интервала $(5; 10000]$. А это очень трудозатратно. Поэтому решать поставленную задачу мы будем, вычисляя вероятность противоположного события, то есть вероятность того, что сломается пять или меньше ноутбуков. А затем отнимем от общей вероятности, которая равна единице, это значение.

$P_{broken}(X > 5) = 1 - P_{broken}(X \leq 5)$

$P_{broken}(X \leq 5) = P_{broken}(X = 0) + P_{broken}(X = 1) + P_{broken}(X = 2) + P_{broken}(X = 3) + P_{broken}(X = 4) + P_{broken}(X = 5)$

Чтобы рассчитать вероятность того, что за период времени произойдёт  $k$ событий, можно пользоваться следующей формулой:

$P(X = k) = \frac{\lambda^k e^{- \lambda}}{k!}$, где

$k$ - количество сломанных ноутбуков, поочередно будет принимать значение 0,1,2,3,4,5,

$\lambda = n * p = 10000*0.0002 = 2$ - ожидаемое количество поломок ноутбуков за месяц, где

$n = 10000$ - количество ноутбуков,

$p = 0.0002$ - вероятность поломки одного ноутбука

$P_{broken}(X = 0) = \frac{2^0 * 2.71828^{- 2}}{0!} = \frac{1 * 0.13533546530402737}{1} = \frac{0.13533546530402737}{1} = 0.13533546530402737$

$P_{broken}(X = 1) = \frac{2^1 * 2.71828^{- 2}}{1!} = \frac{2 * 0.13533546530402737}{1} = \frac{0.27067093060805475}{1} = 0.27067093060805475$

$P_{broken}(X = 2) = \frac{2^2 * 2.71828^{- 2}}{2!} = \frac{4 * 0.13533546530402737}{2} = \frac{0.5413418612161095}{2} = 0.27067093060805475$

$P_{broken}(X = 3) = \frac{2^3 * 2.71828^{- 2}}{3!} = \frac{8 * 0.13533546530402737}{6} = \frac{1.082683722432219}{6} = 0.1804472870720365$

$P_{broken}(X = 4) = \frac{2^4 * 2.71828^{- 2}}{4!} = \frac{16 * 0.13533546530402737}{24} = \frac{2.165367444864438}{24} = 0.09022364353601825$

$P_{broken}(X = 5) = \frac{2^5 * 2.71828^{- 2}}{5!} = \frac{32 * 0.13533546530402737}{120} = \frac{4.330734889728876}{120} = 0.0360894574144073$

$P_{broken}(X \leq 5) = 0.13533546530402737 + 0.27067093060805475 + 0.27067093060805475 + 0.1804472870720365 + 0.09022364353601825 + 0.0360894574144073 = 0.983437714542599$

$P_{broken}(X > 5) = 1 - 0.983437714542599 = 0.016562285457401016$

Ответ: $P_{broken}(X > 5) = 0.016562285457401016$

Вычислим вероятность поломки с помощью Python

```python
import scipy
import seaborn as sns
from numpy import random

# ожидаемое количество поломок ноутбуков за месяц
lmd = 10000*0.0002

# Вычисление P(X <= 5) с помощью функции кумулятивного распределения (CDF)
prob_at_most_5 = scipy.stats.poisson.cdf(5, lmd)

# Вычисление P(X > 5)
prob_more_than_5 = 1 - prob_at_most_5

print(f"\nВероятность поломки более пяти ноутбуков равна: {prob_more_than_5}")
```

#### 3.3.) Корректное решение второй подзадачи пункта б
Определим математическое ожидание количества поломок. Для распределения Пуассона математическое ожидание (среднее количество поломок) равно параметру $E(X) = \lambda = 10000*0.0002 = 2$.
Таким образом, в среднем компания ожидает 2 поломки ноутбуков в месяц.

Вычислим математическое ожидание общих расходов на поломки за один месяц. Общие расходы линейно зависят от количества сломанных ноутбуков. Стоимость ремонта одного ноутбука 500$.
Математическое ожидание общих расходов:

$E(расходов) = 500 * E(X) = 500 * 2 = 1000$

Ответ: Математическое ожидание общих расходов компании на поломки за один месяц составляет 1000 долларов.


```python
import scipy
import seaborn as sns
from numpy import random

# Стоимость одной поломки
cost_per_breakdown = 500

# ожидаемое количество поломок ноутбуков за месяц
lmd = 10000*0.0002

# математическое ожидание количества поломок для Пуассона равно lambda
expected_num_breakdowns = scipy.stats.poisson.mean(lmd)

# математического ожидания общих расходов
expected_total_cost = expected_num_breakdowns * cost_per_breakdown

print(f"\nМатематического ожидания общих расходов: {expected_total_cost} $")
```
