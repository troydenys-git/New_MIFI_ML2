```python
import pandas as pd
path = r'C:\IDE\python_homework\Пайтон для анализа данных\Модуль 5. Продвинутые приемы Pandas\data\ratings_movies.csv'
rating_movies = pd.read_csv(path)
rating_movies
```

```python
#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
    
year = rating_movies['title'].apply(get_year_release)

```

```python
rating_movies['year_release'] = year

release_1999 = rating_movies[rating_movies['year_release'] == 1999]

release_1999.groupby(by='title')['rating'].mean(numeric_only=True).sort_values(ascending=False)
```
