```python
import pandas as pd
import numpy as np
file_path_1 = r'C:\IDE\python_homework\Пайтон для анализа данных\Модуль 6. Визуализация данных\data\city_relocation_train.csv'
xxx = pd.read_csv(file_path_1)
xxx = xxx['Город, переезд, командировки']
xxx
```

```python
def extract_city(location_str):
    if pd.isna(location_str):
        return np.nan
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 
                  'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 
                  'Красноярск', 'Пермь', 'Воронеж', 'Волгоград']
    # Город всегда идет первым в строке, до первой запятой
    city_name = location_str.split(',')[0].strip()
    
    if city_name == 'Москва':
        return 'Москва'
    elif city_name == 'Санкт-Петербург':
        return 'Санкт-Петербург'
    elif city_name in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'

# Применяем функцию
yyy = xxx.apply(extract_city)
yyy
```
