```python
import pandas as pd 
ufo_data = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv')
ufo_data.head()
```

```python
ufo_data_copy = ufo_data.copy()



ufo_data_copy['Time'] = pd.to_datetime(ufo_data_copy['Time'])
ufo_data_copy['Time']
```

```python
ufo_data_copy['Time'].dt.year.mode()
```

```python

ufo_data_copy['Date'] = ufo_data_copy['Time'].dt.date


result = ufo_data_copy[ufo_data_copy['State'] == 'NV']['Date']
result = result.diff()
result = result.dt.days.mean()
result
```

```python
import pandas as pd 
test_series_1 = pd.Series([
    'Опыт работы 8 лет 3 месяца',
    'Опыт работы 3 года 5 месяцев',
    'Опыт работы 1 год 9 месяцев',
    'Опыт работы 3 месяца',
    'Опыт работы 6 лет'
])
test_series_2 = pd.Series([
    'Опыт работы 5 лет',
    'Опыт работы 5 месяцев',
    'Опыт работы 1 год 1 месяц',
    'Опыт работы 3 месяца',
    'Опыт работы 7 лет'
])
def get_experience(arg):
    year = arg.split()[3][:3]
    if len(arg.split()) > 4:
        month = int(arg.split()[4])
    else:
        month = 0
    if year == 'лет' or year == 'год':
        years_to_month = int(arg.split()[2])*12
        return int(years_to_month+month)
    else:
        return int(arg.split()[2])
     

test_series_2
```

```python
result = test_series_2.apply(get_experience)
result
```
